## Create Dataframe with Dictionary

```python
import pandas as pd
df = pd.DataFrame({'col1' : [1,2,3,4,5],
              'col2' : [45,56,67,34,34],
              'col3': "ajay d k m n".split()
     })
```

```python
df
```

```python
df['col4'] = df['col1']**2
df
```

```python
def test(a) :
    return a**3


df['col5'] = df['col1'].apply(test)
```

```python
df
```

## Create Column based on some other column values (conditional)

```python
def test1(a):
    if a =='d':
        return 'Dance'
    else:
        return 'No-Dance'
    
df['col6'] = df['col3'].apply(test1)
```

```python
df
```

```python
df['col6_6'] = df['col3'].apply(lambda a : 'da' if a == 'd' else 'd-not-found')
```
